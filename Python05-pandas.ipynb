#### JOIN

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': np.random.randn(4)})

df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],
                    'value': np.random.randn(4)})
```

```python
df1
```

```python
df2
```

```python
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
```

```python
pd.merge(df1, df2, on='key')
```

```python
-- show all records from df1
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
-- show all records from df2
SELECT *
FROM df1
RIGHT OUTER JOIN df2
  ON df1.key = df2.key;
```

```python
pd.merge(df1, df2, on='key', how='right')
```

```python
-- show all records from both tables
SELECT *
FROM df1
FULL OUTER JOIN df2
  ON df1.key = df2.key;
```

```python
 pd.merge(df1, df2, on='key', how='outer')
```

### Группировка

```python
SELECT sex, count(*)
FROM tips
GROUP BY sex;
/*
Female     87
Male      157
*/
```

```python
tips.groupby('sex')
```

```python
tips.groupby('sex').size()
```

```python
tips.groupby('sex').count()
```

```python
tips.groupby('sex')['total_bill'].count()
```

```python
SELECT day, AVG(tip), COUNT(*)
FROM tips
GROUP BY day;
/*
Fri   2.734737   19
Sat   2.993103   87
Sun   3.255132   76
Thur  2.771452   62
*/
```

```python
#Аггрегирующая функция
tips.groupby('day').agg({'tip': np.mean, 'day': np.size})
```

```python
tips.groupby('day').agg({'tip': [np.mean, np.min], 'day': np.size})
```

```python
SELECT smoker, day, COUNT(*), AVG(tip)
FROM tips
GROUP BY smoker, day;
/*
smoker day
No     Fri      4  2.812500
       Sat     45  3.102889
       Sun     57  3.167895
       Thur    45  2.673778
Yes    Fri     15  2.714000
       Sat     42  2.875476
       Sun     19  3.516842
       Thur    17  3.030000
*/
```

```python
tips.groupby(['smoker', 'day']).agg({'tip': [np.size, np.mean]})
```
